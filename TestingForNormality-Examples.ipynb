```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import scipy.special as sps
import statsmodels.api as sm
from   scipy.stats import norm
import numpy.random as random
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(10)
nml = rng.normal(0,1,size=10000) #generate data from standard normal distn
nml
```

```python
binedges=np.linspace(-4,4,60)
plt.hist(nml, bins=binedges, density=True)
plt.show()
```

```python
plt.hist(nml, bins=binedges, density=True)
p = norm.pdf(binedges, np.mean(nml), np.std(nml))
plt.plot(binedges, p, 'r', linewidth=2)
plt.show()
```

```python
fig = sm.qqplot(nml, dist=norm, loc=0,scale=2)
plt.show()
```

```python
stats.skew(nml) #skewness has to be close to zero
```

```python
stats.kurtosis(nml, fisher=False)  #kurtosis has to be close to 3
```

```python
JB,pvalue=stats.jarque_bera(nml)
print('JB = ', JB)
print('pvalue = ', pvalue)  #large pvalue means the null hypothesis is true
```

```python
KS,pvalue=stats.kstest(nml,'norm')
print('KS = ', KS)
print('pvalue = ', pvalue) #large pvalue means the null hypothesis is true
```
