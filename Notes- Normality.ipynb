#   Moments: 

#### The sample mean and variance are $\overline x=\frac{1}{n}\, \sum_{i=1}^n x_i$ and $\sigma^2=\frac{1}{n}\sum_{i=1}^n (x_i-\overline x)^2$

#### The third and fourth moments are $\mu_3=\frac{1}{n}\sum_{i=1}^n(x_i-\overline x)^3$ and $\mu_4=\frac{1}{n}\sum_{i=1}^n(x_i-\overline x)^4$

# Skewness:              
 
 S $=\displaystyle \frac{\mu_3}{\sigma^{3}}$
 
 #### Skewness gives a measure of how symmetric the observations are about the mean. For a normal distribution the skewness is zero. A distribution skewed to the right has postive skewness and a distribution skewed to the left has negative skewness

# Kurtosis:         

K $ = \displaystyle\frac{\mu_4}{\sigma^4}$

#### Kurtosis gives a measure of the thickness in the tails of a distribution. For a normal distribution the kurtosis is 3. A thick tailed distribution has kurtosis that exceeds 3.

# Checking validity of normality assumption:

### Histogram and Q-Q plot are visual tools to check for normality 

## Consider testing the hypothesis 

#### $H_0$ : The distribution is normal, that is, skewness is zero and kurtosis equals 3

#### versus

#### $H_1$ : The distribution is not-normal

## The Jarque-Bera test statistic is 

#### JB $ = $ (#obs) . $\bigl[\frac{S^2}{6} + \frac{(K-3)^2}{24} \bigr]$

#### If JB is large we reject the null hypothesis 

## The Kolmogorov-Smirnov Test Statistic

#### The observed cumulative distribution function is

###  $F_{obs}(x)=\frac{ \# \;\; data\;\; \leq \;\; x}{total}$

#### $H_0$ : The distribution is $F_{0}(x)$ $\quad$  versus   $\quad$  $H_1$ : The distribution is not $F_{0}(x)$ 
 

### The K-S test statistic is 

###  KS $=$ max $|F_{obs}(x)\; - \; F_{0} (x)|$

### We reject the null hypothesis if KS is large

### The blue line is $F_{obs}(x)$ and the red line is $F_0(x)$
